# Data import and cleaning : 

```python
skillcraft = pd.read_csv("SkillCraft1_Dataset.csv")
skillcraft
```

## Deleting unexpected character "?" :

```python
skillcraft=skillcraft.drop(skillcraft[skillcraft["Age"]=="?"].index,axis=0)
skillcraft=skillcraft.drop(skillcraft[skillcraft["HoursPerWeek"]=="?"].index,axis=0)
skillcraft=skillcraft.drop(skillcraft[skillcraft["TotalHours"]=="?"].index,axis=0)
```

```python
skillcraft.dtypes
```

We can see that there are three columns that should be integer but are not so we need to transform them to get a better visualisation for later.

## Changing types from object to integer : 

```python
skillcraft["Age"]=skillcraft["Age"].astype(int)
skillcraft["HoursPerWeek"]=skillcraft["HoursPerWeek"].astype(int)
skillcraft["TotalHours"]=skillcraft["TotalHours"].astype(int)
```

## Searching for null values

```python
skillcraft["Age"]=skillcraft["Age"].astype(int)
skillcraft["HoursPerWeek"]=skillcraft["HoursPerWeek"].astype(int)
skillcraft["TotalHours"]=skillcraft["TotalHours"].astype(int)
```

```python
sc=skillcraft
league=skillcraft.LeagueIndex
```

```python
# feature cleanup
drops = ["GameID", "LeagueIndex","Age","TotalHours","ActionsInPAC","UniqueUnitsMade","ComplexUnitsMade","ComplexAbilitiesUsed"]
#drops = ["GameID","LeagueIndex"]
sc.drop(drops, axis = 1, inplace = True)
sc.head()
```

```python
league
```

```python
#columns=['Age', 'HoursPerWeek', 'TotalHours', 'APM', 'SelectByHotkeys', 'AssignToHotkeys', 'UniqueHotkeys', 'MinimapAttacks', 'MinimapRightClicks', 'NumberOfPACs', 'GapBetweenPACs', 'ActionLatency', 'ActionsInPAC', 'TotalMapExplored', 'WorkersMade', 'UniqueUnitsMade', 'ComplexUnitsMade', 'ComplexAbilitiesUsed']
       
scaler = MinMaxScaler()
scaler.fit(sc)
x_scaled=scaler.transform(sc)
skill_scale=pd.DataFrame(x_scaled, columns=['HoursPerWeek','APM', 'SelectByHotkeys',
       'AssignToHotkeys', 'UniqueHotkeys', 'MinimapAttacks',
       'MinimapRightClicks', 'NumberOfPACs', 'GapBetweenPACs', 'ActionLatency', 'TotalMapExplored','WorkersMade'])
skill_scale
```

# Splitting Dataset

```python
  y = league
x_train, x_test, y_train, y_test = train_test_split(skill_scale, y, test_size = 0.2, random_state=123,stratify = league)
```

```python
print(f"Training dataset shape: {x_train.shape}")
print(f"Test dataset shape: {x_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Test target shape: {y_test.shape}")
```

x_train.drop("LeagueIndex", axis = 1, inplace = True)
x_test.drop("LeagueIndex", axis = 1, inplace = True)

# Fitting Models

### Logistic Regression

```python
# fit the model
reg = LogisticRegression()
# train the model
model_0 = reg.fit(x_train, y_train)
```

```python
# predictions
predictions_0 = reg.predict(x_test)

# view predictions
predictions_0[:10]
```

```python
accuracy_score(y_test, predictions_0)
```

```python
cross_val_score(LogisticRegression(random_state=0), x_train,y_train,cv=5, scoring='accuracy').mean()
```

### Logistic Regression CV

```python
# fit the model
reg = LogisticRegressionCV(cv=5, random_state=0)
# train the model
model_0 = reg.fit(x_train, y_train)
```

```python
# predictions
predictions_0 = reg.predict(x_test)

# view predictions
predictions_0[:10]
```

```python
accuracy_score(y_test, predictions_0)
```

```python
plot_confusion_matrix(model_0, X=x_test, y_true=y_test)
```

### Gaussian Naive Bayes : 

```python
# fit the model
clf = GaussianNB()

# train the model
model_1 = clf.fit(x_train, y_train)
```

```python
# predictions
predictions_1 = clf.predict(x_test)

# view predictions
predictions_1[:10]
```

```python
accuracy_score(y_test, predictions_1)
```

```python
cross_val_score(GaussianNB(), x_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
plot_confusion_matrix(model_1, X=x_test, y_true=y_test)
```

### Random forest : 

Before using gridseach we decided to try to fit a model using n_estimator = 64 and random_state=85

```python
# build the model
clf2 = RandomForestClassifier(n_estimators = 64, random_state = 85)
model_2 = clf2.fit(x_train, y_train)
```

```python
predictions_2 = clf2.predict(x_test)

# view them
predictions_2[:10]
```

```python
accuracy_score(y_test, predictions_2)
```

###  Plotting Scores

```python
model= RandomForestClassifier(random_state=85)
k= np.arange(1,100)
train_score,val_score= validation_curve(model,x_train,y_train,'n_estimators',k,cv=5)

plt.plot(k,val_score.mean(axis=1),label='validation')
plt.plot(k,train_score.mean(axis=1),label='train')

plt.ylabel("score")
plt.xlabel("n_estimators")
plt.legend()
```

Now we used gridsearch to find the best model playing with hyperparameters n_estimator, random_state and criterion

```python
param_grid = {'n_estimators': np.arange(1,100),
             'random_state': np.arange(1,100),
             'criterion': ['gini', 'entropy']}
grid_forest= GridSearchCV(RandomForestClassifier(), param_grid,cv=5)
grid_forest.fit(x_train,y_train)
grid_forest.best_score_
```

```python
#grid.best_params_
best_mod=grid_forest.best_estimator_
best_mod.score(x_test,y_test)
grid.best_score_
best_mod
grid.best_params_
```

KNN :

Before using gridsearch we tryed to implement an algorithm that seach the best score using the variation of the hyperparameter n_neighbors from 1 too 1000

```python
model= KNeighborsClassifier(n_neighbors=1,)
model.fit(x_train,y_train)
k=1
scor=model.score(x_test,y_test)
for i in range(2,1000):
    model= KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    if(model.score(x_test,y_test)>scor):
        k=i
        scor=model.score(x_test,y_test)
print(scor,k)
```

### Ploting scores 

```python
model= KNeighborsClassifier()
k= np.arange(1,50)
train_score,val_score= validation_curve(model,x_train,y_train,'n_neighbors',k,cv=5)

plt.plot(k,val_score.mean(axis=1),label='validation')
plt.plot(k,train_score.mean(axis=1),label='train')

plt.ylabel("score")
plt.xlabel("n_neighbors")
plt.legend()
```

optimizing KNN :

```python
param_grid = {'n_neighbors': np.arange(1,100),
             'metric': ['euclidean','manhattan']}
grid= GridSearchCV(KNeighborsClassifier(), param_grid,cv=5)
grid.fit(x_train,y_train)
grid.best_score_
```

score of grid search on KNN : 

```python
best_mod_knn=grid.best_estimator_
best_mod_knn.score(x_test,y_test)
```

confusion Matrix :

```python
plot_confusion_matrix(best_mod_knn, X=x_test, y_true=y_test)
```
